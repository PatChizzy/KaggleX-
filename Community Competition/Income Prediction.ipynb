## Data cleaning

```python
# drop the migration information columms since their missing values percentage is too large
clean_df = df.drop(['MIGMTR1', 'MIGMTR3', 'MIGMTR4', 'MIGSUN', 'ID'], axis = 1)
test_df = test.drop(['MIGMTR1', 'MIGMTR3', 'MIGMTR4', 'MIGSUN'], axis = 1)
```

```python
# fill remaining columns with missing values with mode 
col = ['GRINST', 'PEFNTVTY', 'PEMNTVTY', 'PENATVTY']

for i in col:
    clean_df[i] = clean_df[i].fillna(clean_df[i].mode()[0])
    test_df[i] = test_df[i].fillna(test_df[i].mode()[0])
```

```python
clean_df.AREORGN.unique()
```

```python
# select only adults
clean_df = clean_df[clean_df['AAGE'] > 17]
test_df = test_df[test_df['AAGE'] > 17]
```

# Preprocessing

```python
clean_df.AHGA.unique()
```

```python
# Create a dictionary to map sch  names to integers
sch_dict = {' Less than 1st grade' : 0,
              ' 1st 2nd 3rd or 4th grade': 1, 
              ' 5th or 6th grade': 2, 
              ' 7th and 8th grade': 3, 
              ' 9th grade': 4, 
              ' 10th grade': 5,
              ' 11th grade': 6, 
              ' 12th grade no diploma': 7, 
              ' High school graduate': 8, 
              ' Some college but no degree': 9, 
              ' Associates degree-academic program': 10,
              ' Associates degree-occup /vocational': 11, 
              ' Bachelors degree(BA AB BS)': 12, 
              ' Masters degree(MA MS MEng MEd MSW MBA)': 13, 
              ' Prof school degree (MD DDS DVM LLB JD)': 14, 
              ' Doctorate degree(PhD EdD)': 15}

clean_df['AHGA'] = clean_df['AHGA'].map(sch_dict)
test_df['AHGA'] = test_df['AHGA'].map(sch_dict)
```

```python
wrk_dict = {' Not in universe' : 0,
              ' Private': 1, 
              ' Self-employed-not incorporated': 2, 
              ' Federal government': 5, 
              ' State government': 4, 
              ' Local government': 3,
              ' Without pay' : 2,
              ' Never worked' : 2,
              ' Self-employed-incorporated': 3}

clean_df['ACLSWKR'] = clean_df['ACLSWKR'].map(wrk_dict)
test_df['ACLSWKR'] = test_df['ACLSWKR'].map(wrk_dict)
```

```python
ahs_dict = {' Not in universe' : 0,
              ' College or university': 2, 
              ' High school': 1}

clean_df['AHSCOL'] = clean_df['AHSCOL'].map(ahs_dict)
test_df['AHSCOL'] = test_df['AHSCOL'].map(ahs_dict)
```

```python
mar_dict = {' Never married' : 0,
              ' Married-civilian spouse present': 1, 
              ' Married-A F spouse present': 1,
              ' Married-spouse absent': 2, 
              ' Separated': 3, 
              ' Widowed': 4, 
              ' Divorced': 4}

clean_df['AMARITL'] = clean_df['AMARITL'].map(mar_dict)
test_df['AMARITL'] = test_df['AMARITL'].map(mar_dict)
```

```python
from sklearn.preprocessing import LabelEncoder

clean_df = clean_df.apply(LabelEncoder().fit_transform)
test_df = test_df.apply(LabelEncoder().fit_transform)
```

# Feature selection

```python
clean_df.columns
```

```python
df = df.reset_index()
```

```python
X = clean_df.drop(['TARGET'], axis =1)
y = clean_df['TARGET']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train.shape, y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Model

```python
from sklearn.tree import DecisionTreeClassifier

DT_clf = DecisionTreeClassifier()
DT_clf = DT_clf.fit(X_train,y_train)

y_pred_DT = DT_clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print('Confusion Matrix: \n',confusion_matrix(y_test,y_pred_DT))
print(classification_report(y_test,y_pred_DT))
print('Decision tree model Accuracy: ', accuracy_score(y_test, y_pred_DT))
```

```python
from sklearn.naive_bayes import GaussianNB

NB_clf = DecisionTreeClassifier()
NB_clf = NB_clf.fit(X_train,y_train)

y_pred_NB = NB_clf.predict(X_test)
```

```python
print('Confusion Matrix: \n',confusion_matrix(y_test,y_pred_NB))
print(classification_report(y_test,y_pred_NB))
print('Naive Bayes Accuracy : ', accuracy_score(y_test, y_pred_NB))
```

```python
from sklearn.linear_model import LogisticRegression

LR_clf = DecisionTreeClassifier()
LR_clf = LR_clf.fit(X_train,y_train)

y_pred_LR = LR_clf.predict(X_test)
```

```python
print('Confusion Matrix: \n',confusion_matrix(y_test,y_pred_LR))
print(classification_report(y_test,y_pred_LR))
print('Logistic Regression Accuracy : ', accuracy_score(y_test, y_pred_LR))
```
